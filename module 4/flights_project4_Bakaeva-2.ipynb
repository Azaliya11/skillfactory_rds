```python
import pandas as pd
import math
from datetime import datetime, date, time
```

```python
flights = pd.read_csv('flights.csv')
flights
```

```python
flights.info()
```

```python
#удалим повторяющиеся колонки
flights.drop(['flight_id.1','fare_conditions.1'],axis=1,inplace=True)
```

```python
#заполним пропуски нулями
flights.sold_tickets = flights.sold_tickets.fillna(0)
flights.total_amount_tickets = flights.total_amount_tickets.fillna(0)
```

```python
#рассчитаем % заполняемости салона - разделим количество проданных билетов на 
#общее количество мест в каждом классе самолета и умножим на 100

flights['aircraft_occupancy'] = round(flights['sold_tickets'] / flights['seat'] * 100)
flights.sample(7)
```

```python
#приведем данные в колонках 'scheduled_departure','scheduled_arrival','actual_departure','actual_arrival' к формату даты

def to_date(x):
    x = datetime.strptime(x,'%Y-%m-%dT%H:%M:%SZ')
    return x

for column in ['scheduled_departure','scheduled_arrival','actual_departure','actual_arrival']:
    flights[column] = flights[column].apply(lambda x:to_date(x))
flights.sample(7)
```

```python
#рассчитаем длительность полета в минутах
flights['duration'] = flights['actual_arrival'] - flights['actual_departure']
flights['duration'] = flights['duration'].apply(lambda x:x.seconds / 60)
flights.sample(7)
```

```python
#рассчитаем задержку вылета рейса в минутах
flights['delay'] = flights['actual_departure'] - flights['scheduled_departure']
flights['delay'] = flights['delay'].apply(lambda x:x.seconds / 60)
flights.sample(7)
```

```python
#сгруппируем данные по рейсам и проанализируем стоимость билетов, заполняемость салона, длительность полета и задержку
flights.groupby(['flight_no','model'])['total_amount_tickets','aircraft_occupancy','duration', 'delay'].mean()
```

```python
#вынесем в отдельную колонку день недели
flights['day_of_week'] = flights['scheduled_departure'].apply(lambda x:x.weekday())
flights.sample(7)
```

```python
flights.groupby(['flight_no','model','day_of_week'])['total_amount_tickets','aircraft_occupancy','duration', 'delay'].mean().sort_values(by='total_amount_tickets')
```

## РАСХОД ТОПЛИВА разных самолетов
(согласно данным vpolete.online)
Boeing 737-300 - 2,4 тыс. кг/ч
Sukhoi Superjet 100 - 1,7 тыс. кг/ч

```python
#укажем для каждого рейса актуальную цену топлива
prices = {1: 41435, 2:39553, 12:47101}
flights['fuel'] = flights.actual_departure.apply(lambda x:prices[x.month])
flights
```

```python
#рассчитаем для каждого рейса сумму, затраченную на топливо
flights['waste_fuel'] = 0
for i in flights.index:
    if flights.model.iloc[i] =='Boeing 737-300':
        flights.waste_fuel.iloc[i] = 2.4 * (math.ceil(flights.duration.iloc[i]/60)) * flights.fuel.iloc[i]
    else:
        flights.waste_fuel.iloc[i] = 1.7 * (math.ceil(flights.duration.iloc[i]/60)) * flights.fuel.iloc[i]
        
flights
```

```python
#рассчитаем примерную прибыльность каждого рейса
flights['profit'] = flights['total_amount_tickets'] - flights['waste_fuel']
flights.sample(7)
```

```python
flights.groupby(['flight_no','model','day_of_week'])['total_amount_tickets','aircraft_occupancy','profit'].mean().sort_values(by='total_amount_tickets')
```
